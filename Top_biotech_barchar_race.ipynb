```python
import bar_chart_race as bcr
import pandas as pd
```

```python
pharma_size_df = pd.read_csv('source.csv', index_col=False)
pharma_size_df.drop(columns=['Rank'], inplace=True)
pharma_size_df.reset_index(drop=True, inplace=True)
pharma_size_df
```

```python
for year in range(2012,2022):
    pharma_size_df[str(year)] = pharma_size_df[str(year)].str.replace(r"\[.*\]","")
    pharma_size_df[str(year)] = pd.to_numeric(pharma_size_df[str(year)], errors='coerce')
```

```python
# pharma_size_df.columns = ['Company','2021-01-01','2020-01-01','2019-01-01','2018-01-01','2017-01-01','2016-01-01','2015-01-01','2014-01-01','2013-01-01','2012-01-01']
```

```python
pharma_size_df.set_index('Company', inplace=True)
pharma_size_df = pharma_size_df.fillna(0)
pharma_size_df
```

```python
#
pharma_size_t_df = pharma_size_df.T
pharma_size_t_df = pharma_size_t_df.fillna(0)
pharma_size_t_df.sort_index(inplace = True)
pharma_size_t_df
```

```python
import bar_chart_race as bcr
bcr.bar_chart_race(
    df=pharma_size_t_df,
    filename='Top_biotech.gif',
    orientation='h',
    sort='desc',
    n_bars=20,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=20,
    period_length=2000,
    interpolate_period=False,
    label_bars=True,
    bar_size=.90,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va':'center'},
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
             's': 'Animator: P. Yeung\nSource: https://en.wikipedia.org/wiki/List_of_largest_biomedical_companies_by_revenue',
             'ha': 'right', 'size': 7},
    figsize=(6.5,5),
    dpi=288,
    cmap='dark12',
    title='Largest biomedical companies by revenue in Billions USD',
    title_size='',
    bar_label_size=7,
    tick_label_size=5,
    shared_fontdict={'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7})
```
